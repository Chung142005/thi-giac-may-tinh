```python
# khai báo thư viện
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

Yêu	cầu	3:	Thực	hiện	thay	đổi	kích	thước	của	ảnh	gốc	theo	tham	số	tỷ	lệ	%	nhập	
vào	từ	bàn	phím.	Nếu	số	nhập	vào	<	100	(thu	nhỏ)	sử	dụng	thuật	toán	
cv2.INTER_NEAREST,	ngược	lại	nếu	số	nhập	vào	>=100	(phóng	to)	sử	dụng	
thuật	toán	cv2.INTER_LINEAR.	Hiển	thị	kết	quả	như	hình	minh	họa	dưới	đây	và	
lưu	ảnh	đã	thay	đổi	vào	thư	mục	Pic/Saves/	trong	đó	MaSV_Scale_In.jpg	nếu	
ảnh	thu	nhỏ,	MSV_Scale_Out.jpg	nếu	phóng	to

```python
tile= float(input("nhập vào tỉ lệ % muốn thay đổi"))
#tính kích thước ảnh
h_new=int(h*tile/100)
w_new=int(w*tile/100)
if(tile<100):

    anh_resize=cv2.resize(anh,(w_new,h_new),cv2.INTER_NEAREST)
    #hiển thị
    plt.subplot(1,2,1)
    plt.imshow(anh)
    plt.title("ảnh gốc")

    plt.subplot(1,2,2)
    plt.imshow(anh_resize)
    plt.title("ảnh sau resize")

    plt.show()

    #lưu ảnh
    cv2.imwrite('Pic/Saves/2321050024_Scale_In.jpg',cv2.cvtColor(anh_resize,cv2.COLOR_RGB2BGR))
else:
    

    anh_resize=cv2.resize(anh,(w_new,h_new),cv2.INTER_LINEAR)
    #hiển thị
    plt.subplot(1,2,1)
    plt.imshow(anh)
    plt.title("ảnh gốc")

    plt.subplot(1,2,2)
    plt.imshow(anh_resize)
    plt.title("ảnh sau resize")

    plt.show()

    #lưu ảnh
    cv2.imwrite('Pic/Saves/2321050024_Scale_Out.jpg',cv2.cvtColor(anh_resize,cv2.COLOR_RGB2BGR))


```

kích	thước	nếu	ngày	sinh	<=15	lật	ảnh	theo	trục	X,	ngược	lại	lật	ảnh	theo

là	số	thứ	tự	tăng	dần.		
3. Đọc	ảnh	màu	với	hệ	màu	RGB	và	Hiển	thị	toàn	bộ	ảnh	đã	xử	lý	trong	thư	
mục	Pic_New	ra	màn	hình.	

```python
#khai báo thư viện
import os 
import shutil
import pathlib
import matplotlib.pyplot as plt

#đường dẫn
path_original = 'Pic/Pic_original/'

#lấy danh sách tên file có trong thư mục
pic_fish = list(pathlib.Path(path_original).glob('*'))

path_new = 'Pic/Pic_new/'
for i, filename in enumerate(pic_fish):
    # Đọc ảnh
    img = cv2.imread(str(filename))
    
    # Resize ảnh về 196x128
    img_resized = cv2.resize(img, (196, 128))

    #lật ảnh
    img_flip =cv2.flip(img_resized,0)

    #tạo tên
    new_filename = f'2321050024_{i+1}.png'

    #lưu ảnh
    # Lưu ảnh vào thư mục mới
    cv2.imwrite(os.path.join(path_new, new_filename), img_flip)
    print(f"Đã xử lý và lưu: {new_filename}")
```

```python
pic_fish = list(pathlib.Path(path_new).glob('*'))

#Thiết lập số hàng, số cột hiển thị ảnh
cot = 4
hang = len(pic_fish)//cot + 1

#Hiển thị nhiều ảnh sử dụng Matplotlib:
plt.figure(figsize=(4*cot,4*hang))
for i, filename in enumerate(pic_fish):
    # Đọc ảnh
    img = cv2.imread(str(filename))
    img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

    #Thiết lập khung hiển thị:
    plt.subplot(hang,cot,i+1)
    plt.imshow(img)
    
    #lấy tên file hiển thị
    plt.title(str(filename).split("/")[-1])
    plt.grid(ls='--')
    plt.axis('on')
plt.show()
```
